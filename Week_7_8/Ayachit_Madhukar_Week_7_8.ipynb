```python
# eliminating records with null Internal IDs
df_Q6_new=df_q6[df_q6.Internal_ID.isnull().values != True ]
df_Q6_new
```

```python
# grouping to identify breakup and it also helpd to identify incorrect country enteries which in later section i cleaned up

df_main_new.groupby('COUNTRY')['Internal_ID'].count()
```

```python
# cleaning and replacing country names for consistent  names

##USA 
df_main_new['COUNTRY']=df_main_new['COUNTRY'].str.strip().str.lower().replace(to_replace =['united states of america',
                                                        'united ststes',
                                                        'united staes',
                                                        'usausausa',
                                                        'u.s.a.',
                                                        'us',
                                                        'usa ','alaska',
                                                        'i pretend to be from canada, but i am really from the united states.',
                                                        'usas','ussa',
                                                        'usa? hard to tell anymore..', "'merica", 'pittsburgh',
                                                        'united state','ahem....amerca','new york','california','new jersey','north carolina',
                                                        'murica','unites states','usaa','u s', 'u.s.','united stated',
                                                        'usa usa usa!!!!','usa! usa! usa!', 'u s a' 'usaa' ,'united statss'  ,'united statss',                           
                                                        'united states','united sates','unied states','u s a', 'united statea','america',
                                                        'the united states of america', 'unite states','us of a', 'the united states','n. america','unhinged states'], 
                            value ="usa")

## UK
df_main_new['COUNTRY']=df_main_new['COUNTRY'].str.strip().str.lower().replace(to_replace =['endland','england','u.k.','uk', 'united kingdom'],value ="UK")
df_main_new['COUNTRY']=df_main_new['COUNTRY'].str.strip().str.lower().replace(to_replace =['can','canada','canada`'],value ="Canada")

## Garbage names 
garbage=['narnia',
       'subscribe to dm4uz3 on youtube', "i don't know anymore",'a','insanity lately','earth',
       'fear and loathing','ud','trumpistan','murrika','soviet canuckistan','atlantis','cascadia','canae']

## netherland
df_main_new['COUNTRY']=df_main_new['COUNTRY'].str.strip().str.lower().replace(to_replace =['netherlands','the netherlands'],value ='netherlands')
## replacing garbage names with NA
df_main_new['COUNTRY']=df_main_new['COUNTRY'].str.strip().str.lower().replace(to_replace =garbage,value =np.NAN)

# repeated it multiple times and kept addding icorrect names for appropriate replacement  to get clean country name
df_main_new['COUNTRY'].str.lower().unique()
```

```python
# media has multiple columns which can be clubbed into one 

media=['Internal_ID', 'MEDIA_Daily_Dish', 'MEDIA_Science',
       'MEDIA_ESPN', 'MEDIA_Yahoo','media']
conditions = [
    (df_main_new['MEDIA_Daily_Dish'] == 1.0),
    (df_main_new['MEDIA_Science'] == 1.0),
    (df_main_new['MEDIA_ESPN'] ==1.0),
    (df_main_new['MEDIA_Yahoo']==1.0),
    (df_main_new['MEDIA_Daily_Dish'].isna()) & (df_main_new['MEDIA_Science'].isna()) & (df_main_new['MEDIA_ESPN'].isna()) & (df_main_new['MEDIA_Yahoo'].isna())
    ]

    
# create a list of the values we want to assign for each condition
values = ['Daily_Dish', 'Science', 'ESPN', 'Yahoo',np.NaN]

# create a new column and use np.select to assign values to it using our lists as arguments

df_main_new["media"]= np.select(conditions, values)
df_main_new[media]
```

```python
# further droping unused columns and cleaning data for null countries

drop_Columns=['Unnamed','MEDIA_Daily_Dish','MEDIA_Science','MEDIA_ESPN','MEDIA_Yahoo']
df_main_clean=df_main_new.drop(columns=drop_Columns)
df_main_clean=df_main_clean[df_main_clean.COUNTRY.isnull().values != True ]
df_main_clean
```

```python
# combining both the datasets now
#df_main_clean
#df_Q6_new

df_Combined=pd.merge(df_main_clean, df_Q6_new, on='Internal_ID',how='inner')
df_Combined
```
