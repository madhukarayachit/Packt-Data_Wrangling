#### Task-2
Write a list comprehension to generate a second list from the one you just created. The condition of membership in the second list is divisibility by 3.



```python
list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]
```

```python
list_with_divisible_by_3
```

### Task -3 
Calculate the length of these two lists and store the difference in a new variable

```python
length_of_random_list = len(random_number_list)
length_of_3_divisible_list = len(list_with_divisible_by_3)
difference = length_of_random_list - length_of_3_divisible_list
difference
```

#### Task-4
Using a loop, perform steps 2 and 3 and find the difference variable three times.



```python
NUMBER_OF_EXPERIEMENTS = 10
difference_list = []
for i in range(0, NUMBER_OF_EXPERIEMENTS):
    random_number_list = [random.randint(0, LIMIT) for x in range(0, LIMIT)]
    list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]
    # Length of the first list
    length_of_random_list = len(random_number_list)
    # Length of the second list
    length_of_3_divisible_list = len(list_with_divisible_by_3)
    difference = length_of_random_list - length_of_3_divisible_list
    
    difference_list.append(difference)
difference_list
```

#### Task-5
Find the arithmetic mean of these three difference values.


```python
avg_diff = sum(difference_list) / float(len(difference_list))
avg_diff
```

```python
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
```

### Task - 1
Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None.



```python
from itertools import zip_longest
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```
